```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (
    confusion_matrix, accuracy_score, precision_score,
    recall_score, f1_score, roc_auc_score, ConfusionMatrixDisplay
)
```

```python
df = pd.read_csv("college_student_placement_dataset.csv")

df["Placement"] = df["Placement"].map({"No": 0, "Yes": 1})
df["College_ID"] = df["College_ID"].astype("category")
df = pd.get_dummies(df, drop_first=True)
X = df.drop(columns=["Placement"])
y = df["Placement"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=123, stratify=y
)

num_cols = X_train.select_dtypes(include=["int64", "float64"]).columns
scaler = StandardScaler()
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])

print("Baseline Logistic Regression")

baseline_model = LogisticRegression(max_iter=1000)
baseline_model.fit(X_train, y_train)

y_prob_base = baseline_model.predict_proba(X_test)[:, 1]
y_pred_base = (y_prob_base > 0.5).astype(int)

print("Confusion matrix:\n", confusion_matrix(y_test, y_pred_base))
print("Accuracy:", accuracy_score(y_test, y_pred_base))
print("Precision:", precision_score(y_test, y_pred_base))
print("Recall:", recall_score(y_test, y_pred_base))
print("F1:", f1_score(y_test, y_pred_base))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_base))

print("\n Tuned Logistic Regression")

strat_kfold_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)
param_grid = {
    "C": [0.01, 0.1, 1.0, 10.0, 100.0],
    "penalty": ["l1", "l2"],
}

base_log_reg = LogisticRegression(max_iter=1000, solver="liblinear")
grid = GridSearchCV(
    estimator=base_log_reg,
    param_grid=param_grid,
    cv=strat_kfold_cv,
    scoring="f1",
    n_jobs=-1,
    verbose=0
)
grid.fit(X_train, y_train)

print("Best hyperparameters:", grid.best_params_)
print("Best CV F1:", grid.best_score_)

tuned_model = grid.best_estimator_
y_prob_tuned = tuned_model.predict_proba(X_test)[:, 1]
y_pred_tuned = (y_prob_tuned > 0.5).astype(int)

print("Confusion matrix:\n", confusion_matrix(y_test, y_pred_tuned))
print("Accuracy:", accuracy_score(y_test, y_pred_tuned))
print("Precision:", precision_score(y_test, y_pred_tuned))
print("Recall:", recall_score(y_test, y_pred_tuned))
print("F1:", f1_score(y_test, y_pred_tuned))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_tuned))
```

```python
cm = confusion_matrix(y_test, y_pred_tuned, labels=[0, 1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Placed", "Placed"])
fig, ax = plt.subplots(figsize=(4, 4))
disp.plot(cmap="Oranges", values_format="d", ax=ax, colorbar=False)
plt.title("Logistic Regression Confusion Matrix")
plt.tight_layout()
plt.show()
```

```python
feature_names = X_train.columns        
coefs = tuned_model.coef_[0]
coef_importance = pd.Series(coefs, index=feature_names)

top_k = 5
top = coef_importance.sort_values().tail(top_k)

plt.figure(figsize=(8, 5))
top.plot(kind='barh', color="orange")
plt.title("Logistic Regression Feature Importance")
plt.xlabel("Coefficient")
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("college_student_placement_dataset.csv")

df["Placement"] = df["Placement"].map({"No": 0, "Yes": 1})
df = df.drop(columns=["College_ID"])
df = pd.get_dummies(df, drop_first=True)

X = df.drop(columns=["Placement"])
y = df["Placement"]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=123, stratify=y
)

num_cols = X_train.select_dtypes(include=["int64", "float64"]).columns
scaler = StandardScaler()
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])

print("Baseline Gaussian Naive Bayes")

nb_base = GaussianNB()
nb_base.fit(X_train, y_train)

y_prob_nb_base = nb_base.predict_proba(X_test)[:, 1]
y_pred_nb_base = nb_base.predict(X_test)

cm_nb_base = confusion_matrix(y_test, y_pred_nb_base)
acc_nb_base = accuracy_score(y_test, y_pred_nb_base)
prec_nb_base = precision_score(y_test, y_pred_nb_base)
rec_nb_base = recall_score(y_test, y_pred_nb_base)
f1_nb_base = f1_score(y_test, y_pred_nb_base)
roc_auc_nb_base = roc_auc_score(y_test, y_prob_nb_base)

print("Confusion matrix (NB baseline):\n", cm_nb_base)
print("Accuracy (NB baseline):", acc_nb_base)
print("Precision (NB baseline):", prec_nb_base)
print("Recall (NB baseline):", rec_nb_base)
print("F1 (NB baseline):", f1_nb_base)
print("ROC-AUC (NB baseline):", roc_auc_nb_base)

print("\n==== TUNED NAIVE BAYES (GaussianNB + GridSearchCV) ====")

strat_kfold_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)
param_grid_nb = {
    "var_smoothing": np.logspace(-12, -4, 9)
}
nb_model = GaussianNB()
grid_nb = GridSearchCV(
    estimator=nb_model,
    param_grid=param_grid_nb,
    cv=strat_kfold_cv,
    scoring="f1",
    n_jobs=-1,
    verbose=0
)
grid_nb.fit(X_train, y_train)

print("Best params (NB):", grid_nb.best_params_)
print("Best CV F1 (NB):", grid_nb.best_score_)

nb_tuned = grid_nb.best_estimator_
y_prob_nb_tuned = nb_tuned.predict_proba(X_test)[:, 1]
y_pred_nb_tuned = nb_tuned.predict(X_test)

cm_nb_tuned = confusion_matrix(y_test, y_pred_nb_tuned)
acc_nb_tuned = accuracy_score(y_test, y_pred_nb_tuned)
prec_nb_tuned = precision_score(y_test, y_pred_nb_tuned)
rec_nb_tuned = recall_score(y_test, y_pred_nb_tuned)
f1_nb_tuned = f1_score(y_test, y_pred_nb_tuned)
roc_auc_nb_tuned = roc_auc_score(y_test, y_prob_nb_tuned)

print("Confusion matrix (NB tuned):\n", cm_nb_tuned)
print("Accuracy (NB tuned):", acc_nb_tuned)
print("Precision (NB tuned):", prec_nb_tuned)
print("Recall (NB tuned):", rec_nb_tuned)
print("F1 (NB tuned):", f1_nb_tuned)
print("ROC-AUC (NB tuned):", roc_auc_nb_tuned)
```

```python
feature_names = X_train.columns

means = nb_tuned.theta_

mean_diff = np.abs(means[1] - means[0])

importance = pd.Series(mean_diff, index=feature_names)

top_k = 8
top_imp = importance.sort_values().tail(top_k)

plt.figure(figsize=(8, 5))
top_imp.plot(kind="barh", color="green")
plt.xlabel("Mean Difference")
plt.title("Naive Bayes Feature Importance")
plt.tight_layout()
plt.show()
```
